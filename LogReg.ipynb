```python
from sklearn.datasets import fetch_lfw_people
```

```python
face_data = fetch_lfw_people(min_faces_per_person=25)
```

```python
X = face_data.data
Y = face_data.target

print('Input data size :', X.shape)
print('Output data size :', Y.shape)
print('Label names:', face_data.target_names)
```

```python
import numpy as np
import matplotlib.pyplot as plt

unique_targets, counts = np.unique(Y, return_counts=True)

plt.figure(figsize=(20, 18))

plt.barh(np.arange(len(unique_targets)), counts)

plt.yticks(np.arange(len(unique_targets)), [name.replace('_', ' ').replace(' ', '\n') for name in face_data.target_names[unique_targets]])

plt.xlabel('Number of Pictures')
plt.ylabel('Person')
plt.title('Number of Pictures of Each Person in the Dataset')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

unique_targets, counts = np.unique(Y, return_counts=True)

sorted_indices = np.argsort(counts)[::-1]

top_people = face_data.target_names[unique_targets[sorted_indices[:5]]]
top_counts = counts[sorted_indices[:5]]

plt.figure(figsize=(10, 6))
plt.barh(top_people, top_counts)

plt.xlabel('Number of Pictures')
plt.ylabel('Person')
plt.title('Top 5 People with the Most Pictures in the Dataset')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42)
```

```python
X_train.shape, Y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(penalty = 'l2', C = 0.1)
logreg.fit(X_train, Y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
logreg.score(X_train,Y_train)
```

```python
logreg.score(X_test,Y_test)
```
